Comencem amb Beautiful Soup i fem una sol·licitud GET a l'URL de http://quotes.toscrape.com/

```python
url = 'http://quotes.toscrape.com/'
page = requests.get(url)
```

Creem objecte beautifulsoup

```python
soup = BeautifulSoup(page.content,'html.parser')
```

Busquem tots els elements amb la classe CSS 'text i author' 

```python
frases = soup.find_all('span', class_="text")
aut = soup.find_all('small', class_='author')
```

```python
cites = []
autor = []

for i in frases:
        cites.append(i.text)
for i in aut:
        autor.append(i.text)       

print(autor)
```

```python
print(cites)
```

Anem a la pàgina següent

```python
url = 'http://quotes.toscrape.com/page/2/'
page2 = requests.get(url)

soup2 = BeautifulSoup(page2.content,'html.parser')

frases2 = soup2.find_all('span', class_="text")
aut2 = soup2.find_all('small', class_='author')

cites2 = []
autor2 = []

for i in frases2:
        cites2.append(i.text)
for i in aut2:
        autor2.append(i.text)
```

Ho guardem tot a un csv

```python
df_bs = pd.DataFrame({'Autor': autor + autor2, 'Cites': cites + cites})

df_bs
```

Ara ho farem amb Selenium

```python
driver = webdriver.Chrome()
driver.get('https://quotes.toscrape.com/')

content = driver.find_elements(By.XPATH, '//div[@class="quote"]')

cites3 = []
autor3 = []
for element in content:
    cites3.append(element.find_element(By.CLASS_NAME, 'text').text)
    autor3.append(element.find_element(By.CLASS_NAME, 'author').text)
```

Anem a la pàgina següent

```python
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                      '/html/body/div/div[2]/div[1]/nav/ul/li/a')))\
    .click()

content2 = driver.find_elements(By.XPATH, '//div[@class="quote"]')

cites4 = []
autor4 = []
for element in content2:
    cites4.append(element.find_element(By.CLASS_NAME, 'text').text)
    autor4.append(element.find_element(By.CLASS_NAME, 'author').text)

driver.close()
```

Ho guardem tot a un csv

```python
df_se = pd.DataFrame({'Autor': autor3 + autor4, 'Cites': cites3 + cites4})
df_se
```

```python
df_se.to_csv('QuotesSele.csv')
```

Passem a fer una recerca a la pàgina de la Wikipedia i començarem amb Beautiful soup

```python
url = "https://es.wikipedia.org/wiki/Los_Simpson"
response = requests.get(url)
```

Creem objecte beautifulsoup i busquem la taula de les temporades

```python
soup = BeautifulSoup(response.content, "html.parser")

Simpson = soup.find("table", class_="wikitable plainrowheaders")

Temporada = []
Episodis = []
Primera_emissio = []
Ultima_emissio = []
Audiencia_mitjana = []
Posicio = []

   
for fila in Simpson.find_all("tr")[1:]:
    columnes = fila.find_all("td")
    if len(columnes) >= 6:
        Temp = columnes[0].text.strip()
        Epi = columnes[1].text.strip()
        Primera = columnes[2].text.strip()
        Ultima = columnes[3].text.strip()
        Audiencia = columnes[4].text.strip()
        Pos = columnes[5].text.strip()

        Temporada.append(Temp)
        Episodis.append(Epi)
        Primera_emissio.append(Primera)
        Ultima_emissio.append(Ultima)
        Audiencia_mitjana.append(Audiencia)
        Posicio.append(Pos)

df_simpson = pd.DataFrame({
    "Temporada": Temporada,
    "Episodis": Episodis,
    "Primera_emissio": Primera_emissio,
    "Ultima_emissio": Ultima_emissio,
    "Audiencia_mitjana": Audiencia_mitjana,
    "Posicio": Posicio
})
df_simpson
```

Ho guardem a un csv

```python
df_simpson.to_csv("simpsonBs.csv")
```

Fem la mateixa taula de les temporades dels Simpsons amb Selenium

```python
driver = webdriver.Chrome()
driver.get('https://es.wikipedia.org/wiki/Los_Simpson')

tabla = driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/div[31]/div/table/tbody')

filas = tabla.find_elements_by_tag_name('tr')

Temporada2 = []
Episodis2 = []
Primera_emissio2 = []
Ultima_emissio2 = []
Audiencia_mitjana2 = []
Posicio2 = []

for fila in filas:
    celdas = fila.find_elements_by_tag_name('td')
    if len(celdas) >= 6:  
        Temporada2.append(celdas[0].text)
        Episodis2.append(celdas[1].text)
        Primera_emissio2.append(celdas[2].text)
        Ultima_emissio2.append(celdas[3].text)
        Audiencia_mitjana2.append(celdas[4].text)
        Posicio2.append(celdas[5].text)
```

```python
df_masi = pd.DataFrame({
    "Temporada": Temporada2,
    "Episodis": Episodis2,
    "Primera_emissio": Primera_emissio2,
    "Ultima_emissio": Ultima_emissio2,
    "Audiencia_mitjana": Audiencia_mitjana2,
    "Posicio": Posicio2
})

df_masi

```

```python
driver.close()
```

```python
df_masi.to_csv("simpsonsMasi.csv", index=False)
```
